```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#reading data
dfX = pd.read_csv("./Training Data/Linear_X_Train.csv")
dfY = pd.read_csv("./Training Data/Linear_Y_Train.csv")
X = dfX.values
Y = dfY.values
X = X.reshape((-1,1))
Y = Y.reshape((-1,1))
X = (X - X.mean())/X.std()
plt.scatter(X,Y, color='g')
plt.show()
```

```python
#training data
model = LinearRegression()
model.fit(X,Y)
```

```python
#predicting data
df = pd.read_csv("./Testing Data/Linear_X_Test.csv")
x = df.values
x_test = x.reshape((-1,1))
x_test = (x_test - x_test.mean()) / x.std()
y_test = model.predict(x)
y_test
```

```python
#chechking accuracy
model.score(x_test,y_test)
```

```python
#visualizing the output
plt.scatter(X,Y,label='data', color='g')
plt.plot(x_test,y_test,color='r',label='prediction')
plt.legend()
plt.show()
```
